```python
path='pydata-book/datasets/bitly_usagov/example.txt'
open(path).readline()
```

```python
import json
records=[json.loads(line) for line in open(path)]
records[0]
```

```python
time_zones=[rec['tz'] for rec in records if 'tz' in rec]
time_zones[0:10]
```

```python
from collections import defaultdict
def get_count(sequence):
    counts=defaultdict(int)
    for cell in sequence:
        counts[cell]+=1
    return counts
```

```python
counts=get_count(time_zones)
print(counts)
```

```python
def top_counts(counts_dict,top=10):
    value_pairs=[(count,tz) for tz,count in counts_dict.items()]
    value_pairs.sort()
    return value_pairs[-top:]
top_counts(counts)
```

```python
from collections import Counter
common_count=Counter(time_zones)
common_count.most_common(10)
```

```python
frame['a'][50]
```

```python
results=Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
cframe=frame[frame.a.notnull()]
cframe
```

```python
operating_system=np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
operating_system[:5]
```

```python
by_tz_os=cframe.groupby(['tz',operating_system])
```

```python
agg_counts=by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
indexer=agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset=agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh',stacked=True)
plt.show()
```

```python
normed_subset=count_subset.div(count_subset.sum(1),axis=0)
normed_subset.plot(kind='barh',stacked=True)
plt.show()
```
